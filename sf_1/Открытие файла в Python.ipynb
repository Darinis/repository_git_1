```python
categories = []
for item in data_list:    
    category = item['category']
    categories.append(category)
print (categories)
```

```python
import collections
c = collections.Counter()
for category in categories:
    c[category] += 1
print(c)
print(sum(c.values()))
```

```python
for cnt in c.values():
    print(cnt)
```

```python
table_clients = []
for item in data_list:
    client_id = item['client_id']
    category = item['category']
    if category == 'table':
        table_clients.append(client_id)
print (table_clients)
```

```python
import collections
q = collections.Counter()
for table_client in table_clients:
    q[table_client] += 1
print(q)    
```
